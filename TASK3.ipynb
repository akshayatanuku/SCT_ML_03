```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
# 1. Load the Image Dataset (Built-in for easy execution)
digits = datasets.load_digits()
# 2. Preprocess: Flatten the images
# Images are 8x8 pixels. We turn them into a single row of 64 numbers.
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
# 3. Filter for two classes (Binary Classification: Cat vs Dog proxy)
# We will only look at images of '0' and '1'
images_filter = (digits.target == 0) | (digits.target == 1)
X = data[images_filter]
y = digits.target[images_filter]
# 4. Split into Training (70%) and Testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# 5. Create the SVM Model (using a Linear Kernel for clear boundaries)
model = SVC(kernel='linear')
model.fit(X_train, y_train)
# 6. Make Predictions
y_pred = model.predict(X_test)
# 7. Print Results
print(f"SVM Classification Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nDetailed Performance Report:")
print(classification_report(y_test, y_pred))
# 8. Visualize the Output
plt.imshow(X_test[0].reshape(8,8), cmap='gray')
label = "Cat (0)" if y_pred[0] == 0 else "Dog (1)"
plt.title(f"AI Prediction: {label}")
plt.show()
```
